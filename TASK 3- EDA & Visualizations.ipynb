# Zomato Bangalore Dataset Analysis & Visualization

##  Exploratory Data Analysis (EDA)

### Import Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Dataset

```python
df = pd.read_csv(r"C:\Users\abhir\cleaned_zomato_dataset")
df.head(5)
```

### ðŸ”¹ 1. Distribution of Ratings

---
**Step:** Plot the distribution of restaurant ratings.

**Observation:** Ratings are concentrated between 3.5 and 4.5, with fewer restaurants rated very low.

**Business Value:** Indicates that customer satisfaction is generally high, with intense competition among well-rated businesses.


### ðŸ”¹ 34. Cost vs Rating Bucketed Bar Plot

---
**Step:**  
Bar chart showing average restaurant cost grouped by rating buckets.

**Observation:**  
Restaurants rated 0â€“2 have the highest average cost, while those rated 2â€“3 and 3â€“4 cost less. Restaurants rated 4â€“5 are moderately expensive, but not the highest.

**Business Value:**  
Highlights how pricing varies with rating. Helps identify if higher-priced restaurants tend to have lower ratings, which can guide pricing strategies and quality improvement efforts.



```python
plt.figure(figsize=(10,5))
df['rating_bucket'] = pd.cut(df['rate'], bins=[0, 2, 3, 4, 5], labels=['0-2','2-3','3-4','4-5'])
cost_by_rating = df.groupby('rating_bucket', observed=True)['cost'].mean()
cost_by_rating.plot(kind='bar', color='#9370DB')
plt.title('Average Cost by Rating Bucket')
plt.xlabel('Rating Bucket')
plt.ylabel('Average Cost');
```

### ðŸ”¹ 35. Correlation Heatmap of Numeric Variables
---

**Step:**  
Correlation heatmap of numeric variables ('rate', 'cost', 'votes').

**Observation:**

**Rating & Votes (0.45):** Restaurants with higher ratings tend to get more votes (indicating popularity).

None of these relationships are extremely strong, but all show that these factors tend to go up together to some extent.
